```python
for i, col in enumerate(df_master.columns[:-1]):
    plt.figure(i)
    plt.ylabel("Frequency", fontsize = 15)
    plt.xlabel(col, fontsize = 15)
    plt.hist(df_master[col], color= 'grey', edgecolor = 'b');
```

```python
QR1 = df_master.quantile(0.25)
QR3 = df_master.quantile(0.75)

IQR = QR3 - QR1
((df_master < (QR1 - 1.5 * IQR)) | (df_master > (QR3 + 1.5 * IQR))).any(axis=1)
    
#Print the IQR range 
print(IQR)

print('\n')
print("Outliers in the dataset : (True values in the below table)")
#The value where the "True" exist in below output is the outliers
(df_master < (QR1 - 1.5 * IQR)) | (df_master > (QR3 + 1.5 * IQR))
```

```python
#Remove the outliers( Datapoints outside 25% and 75% Quarters)
df_master = df_master[~((df_master < (QR1 - 1.5 * IQR)) | (df_master > (QR3 + 1.5 * IQR))).any(axis=1)]
df_master.count()
```
